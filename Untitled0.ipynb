```python
X = df.drop(columns=['Traffic_Situation'])
y = df['Traffic_Situation']
```

```python
numeric_features = X.select_dtypes(include=['number']).columns
print(numeric_features)
categorical_features = X.select_dtypes(include=['object']).columns
print(categorical_features)
```

```python
numeric_features = X.select_dtypes(include=['number']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])


preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

pipeline = Pipeline(steps=[('preprocessor', preprocessor)])


X_transformed = pipeline.fit_transform(X)


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y_encoded, test_size=0.2, random_state=42)
```

```python
def train_and_evaluate_models(X_train, X_test, y_train, y_test):
    classifiers = [
        ('lr', LogisticRegression(random_state=42)),
        ('svc', SVC(probability=True, random_state=42)),
        ('dt', DecisionTreeClassifier(random_state=42)),
        ('knn', KNeighborsClassifier()),
        ('xgb', XGBClassifier(random_state=42)),
        ('rf', RandomForestClassifier(random_state=42)),
        ('gb', GradientBoostingClassifier(random_state=42)),
        ('ada', AdaBoostClassifier(random_state=42)),
        ('et', ExtraTreesClassifier(random_state=42)),
        ('bag', BaggingClassifier(random_state=42)),
        ('mlp', MLPClassifier(random_state=42)),
        ('voting', VotingClassifier(
            estimators=[
                ('rf', RandomForestClassifier(random_state=42)),
                ('gb', GradientBoostingClassifier(random_state=42)),
                ('xgb', XGBClassifier(random_state=42))
            ],
            voting='soft'
        )),
        ('stacking', StackingClassifier(
            estimators=[
                ('rf', RandomForestClassifier(random_state=42)),
                ('gb', GradientBoostingClassifier(random_state=42)),
                ('xgb', XGBClassifier(random_state=42))
            ],
            final_estimator=LogisticRegression(random_state=42)
        ))
    ]


    results = {}
    best_score = 0
    best_model = None


    for name, clf in classifiers:
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        score = clf.score(X_test, y_test)

        results[name] = {
            'model': clf,
            'score': score,
            'report': classification_report(y_test, y_pred)
        }

        if score > best_score:
            best_score = score
            best_model = clf

        print(f"Classifier: {name}")
        print(f"Accuracy: {score:.4f}")
        print(classification_report(y_test, y_pred))
        print("-" * 50)


    print("Best Model Saving", best_model)
    joblib.dump(best_model, 'best_model.joblib')

    return results, best_model

results, best_model = train_and_evaluate_models(X_train, X_test, y_train, y_test)
print("Results", results)
print("Best Model", best_model)
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib

def predict_traffic_situation(new_data, model_path='best_model.joblib'):


    model = joblib.load(model_path)


    X_new = pipeline.transform(new_data)


    predictions = model.predict(X_new)
    probabilities = model.predict_proba(X_new)


    decoded_predictions = label_encoder.inverse_transform(predictions)

    return decoded_predictions, probabilities


example_data = pd.DataFrame({
    'Date': [20240101],
    'Time': ['08:00'],
    'DayOfTheWeek': ['Tuesday'],
    'CarCount': [1],
    'BikeCount': [20],
    'BusCount': [10],
    'TruckCount': [5],
    'Total': [36]
})

predictions, probabilities = predict_traffic_situation(example_data)
print("Predicted Traffic Situation:", predictions)
print("Prediction Probabilities:", probabilities)
```
